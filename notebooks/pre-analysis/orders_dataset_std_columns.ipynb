## Preprocessing of file olist_orders_dataset.csv

## Initial Column Description

This file has the problem that it does not have the same number of columns in all rows, there are rows with 10 columns and others with 9.
A cleanup is performed by reading the file flat with the `open()` method. The transformed file is saved in data/pre_interim/orders_dataset_std_columns.csv'.

```python
import pandas as pd
```

```python
with open('../../data/pre_interim/orders_dataset.csv') as f:
    lines = f.readlines()

new_file = list()

new_file.append(lines[0])

for line in lines[1:]:
    #print(len(line.split(',')))
    l = line.split(',')
    if len(l) == 10:
        l.pop(2)
        new_file.append(','.join(l))
    else:
        new_file.append(line)

with open('../../data/pre_interim/orders_dataset_std_columns.csv', 'w') as f:
    for line in new_file:
        f.write(line)

```
