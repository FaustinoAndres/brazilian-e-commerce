# Order reviews dataset 
This dataset includes data about the reviews made by the customers.

After a customer purchases the product from Olist Store a seller gets notified to fulfill that order. Once the customer receives the product, or the estimated delivery date is due, the customer gets a satisfaction survey by email where he can give a note for the purchase experience and write down some comments.

## Initial Column Description


|**Column Title**|**review_id-> str**|**order_id-> str** |**review_score-> int** |**review_comment_title -> str**| **review_comment_message -> str**|**review_creation_date -> timestamp**|**review_answer_timestamp -> timestamp**|
|--|--|--|--|--|--|--|--|
|Description |Primary key - unique review identifier |Primary key - unique order identifier |Note ranging from 1 to 5 given by the customer on a satisfaction survey. |Comment title from the review left by the customer, in Portuguese. |Comment message from the review left by the customer, in Portuguese.| Shows the date in which the satisfaction survey was sent to the customer.|Shows satisfaction survey answer timestamp. |
|Example |8670d52e15e00043ae7de4c01cc2fe06 |b9bf720beb4ab3728760088589c62129 |4 |recomendo	 |aparelho eficiente. no site a marca do aparelho esta impresso como 3desinfector e ao chegar esta com. |2018-05-22 00:00:00 |2018-05-23 16:45:47 |

## Required Libraries

```python
import pandas as pd
```

## Data Preprocessing


```python
review_csv = '../../data/pre_interim/order_reviews_dataset.csv'
review_dataset = pd.read_csv(review_csv)
```

```python
review_dataset
```

```python
review_dataset.rename(columns={'order_id':'order'}, inplace=True)
review_dataset
```

```python
order_id_csv = '../../data/interim/unique_order_id.csv'
order_id = pd.read_csv(order_id_csv, dtype={'order_id':'Int64'})
order_id
```

#### Making merge between `review_dataset`.csv and `order_id`.csv

```python
review_database = pd.merge(review_dataset,order_id, on='order', how='left')
review_database.drop(columns=['order'], inplace=True)
review_database = review_database.reindex(columns=['review_id','review_score','review_comment_title','review_comment_message','review_creation_date','review_answer_timestamp','order_id'])
review_database
```

```python
review_database.describe(include='all')
```

```python
items = len(review_database)
review_id = range(1,items+1)
review_database.insert(0,'id',review_id)
review_database
```

#### Creatign the dataset `review_database.csv`

When you saved the dataset always mark **"index = False"**. Or pandas will add a new column with a consequtive number. This small script is to remove this useless column.

```python
review_database.to_csv('../../data/interim/review_database.csv', index=False)
```
