## Data Preprocessing



```python
df = pd.read_csv(PRODUCTS, dtype={'product_name_lenght': 'Int64', 'product_description_lenght': 'Int64', 'product_photos_qty': 'Int64', 'product_weight_g': 'Int64', 'product_length_cm': 'Int64', 'product_height_cm': 'Int64', 'product_width_cm': 'Int64'})
```

```python
category_name = df['product_category_name'].value_counts()
category_name
```

```python
plt.figure(figsize=(16,8))
ax = category_name.plot.bar(x='product_category_name')
plt.show()
```

```python
with open('../../data/raw/product_category_name_translation.csv') as f:
    lines = f.readlines()

translate = {}

for line in lines[1:]:
    l = line.rstrip('\n').split(',')
    translate[l[0]] = l[1]

len(translate)
```

```python
translate
```

```python
difference_between_category_name_list = set(translate) ^ set(category_name.index)
difference_between_category_name_list
```

```python
translate['pc_gamer'] = 'pc_gamer'
translate['portateis_cozinha_e_preparadores_de_alimentos'] = 'food_preparators'
```

```python
not_nan_category = df[df['product_category_name'].notna()]
not_nan_category.head()
```

```python
with open('../../data/interim/product_category_name_translation.csv') as f:
    lines = f.readlines()

category_id = {}

for line in lines[1:]:
    l = line.rstrip('\n').split(',')
    category_id[l[1]] = int(l[0])

len(category_id)
```

```python
category_id
```

```python
df_product_category_name_id = pd.DataFrame(not_nan_category.apply(lambda row: category_id[row['product_category_name']], axis = 1), columns=['product_category_name_id'], dtype='Int64')
```

```python
df_product_category_name_id.head()
```

```python
not_nan_category = pd.concat([not_nan_category, df_product_category_name_id], axis = 1)
not_nan_category
```

```python
nan_category = df[df['product_category_name'].isna()]
nan_category
```

```python
df_product_category_name_id_nan = pd.DataFrame(nan_category.apply(lambda row: np.nan, axis = 1), columns=['product_category_name_id'], dtype='Int64')
df_product_category_name_id_nan
```

```python
nan_category.head()
```

```python
nan_category = pd.concat([nan_category, df_product_category_name_id_nan], axis = 1)
nan_category
```

```python
df_clean = pd.concat([not_nan_category, nan_category])
```

```python
df_clean
```

```python
df_clean.drop(['product_category_name'], axis = 'columns', inplace = True)
```

```python
columns = list(df_clean.columns)
```

```python
product_category_name_id = columns.pop()
```

```python
columns.insert(1, product_category_name_id)
```

```python
columns
```

```python
df_clean.sort_values(by='product_category_name_id', inplace=True)
df_clean = df_clean.reset_index()
df_clean.drop(['index'], axis = 'columns', inplace = True)
```

```python
df_clean = df_clean[columns]
```

```python
df_clean
```

#### Create the csv 

When you saved the dataset always mark **"index = False"**. Or pandas will add a new column with a consequtive number. This small script is to remove this useless column.

```python
df_clean.to_csv('../../data/interim/products_dataset.csv', index = False, header = True)
```
